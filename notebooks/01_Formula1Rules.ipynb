# Fórmula 1: Regras Básicas

## 🏎️ Estrutura do Campeonato
- A Fórmula 1 é composta por uma temporada anual com **diversas corridas** (normalmente entre 20 e 23).
- Cada corrida distribui pontos aos pilotos e às equipes.

---

## 🏆 Campeonato de Pilotos
- Cada piloto soma pontos ao longo da temporada.
- O piloto com mais pontos ao final da última corrida é declarado campeão mundial.

---

## 🏢 Campeonato de Construtores
- Cada equipe (escuderia) conta com **dois pilotos**.
- Os pontos de ambos os pilotos contam para o campeonato de construtores (a pontuação das sprints também são contabilizadas).
- A equipe com mais pontos no fim da temporada vence.

## ⏱️ Estrutura da Corrida
1. **Qualificação**: define o grid de largada.
2. **Corrida**: define a pontuação.
3. **Resultados**: usados para atualizar a tabela do campeonato.

---

## 🚨 Observações importantes
- **Abandono (DNF)**: pilotos que não completam a corrida não marcam pontos.
- **Penalidades** podem alterar o resultado final.
- Existem regras especiais em caso de corridas encurtadas (ex.: pontuação parcial).
- O sistema de **Sprint Races** (corridas curtas no sábado, mas não acontecem em todos os fins de semanas), começou a ser implementado desde 2021
