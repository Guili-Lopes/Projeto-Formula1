# 🏎️ Projeto: Previsão Probabilística de Resultados na Fórmula 1

> **Estimativa Bayesiana de Probabilidades de Título — Pilotos e Construtores**

---

## 🎯 Objetivo

Desenvolver um sistema que, a cada corrida da temporada de Fórmula 1, estime dinamicamente:

- A **probabilidade de cada piloto** vencer o Campeonato de Pilotos.
- A **probabilidade de cada equipe** vencer o Campeonato de Construtores.

As probabilidades são **atualizadas corrida a corrida**, incorporando os resultados já ocorridos e considerando as corridas restantes da temporada.

---

## 🔬 Metodologia

O projeto é estruturado em **duas abordagens complementares**:

### 1. Modelo Estatístico Bayesiano (principal)

| Componente | Descrição |
|------------|----------|
| **Modelo Plackett–Luce** | Modela rankings completos associando um parâmetro de "habilidade" a cada piloto/equipe |
| **Abordagem Bayesiana Hierárquica** | Trata as habilidades como variáveis aleatórias com distribuições a priori (e.g., Gama, Log-Normal) |
| **Inferência via MCMC** | Amostragem da distribuição a posteriori via Markov Chain Monte Carlo |
| **Power EP (aproximação)** | Expectation Propagation com α-divergence como alternativa escalável ao MCMC |
| **Clustering de Rankings Parciais** | Algoritmo MCMC para agrupar corridas por perfil (pistas rápidas vs. técnicas) usando o Modelo de Mallows |

### 2. Rede Neural (comparação)

- Modelo supervisionado (deep learning) treinado para prever probabilidades de vitória, pódio e posição final.
- Comparação de desempenho entre a abordagem estatística e o aprendizado de máquina.

## 📚 Artigos de Referência

O projeto é fundamentado nos seguintes trabalhos acadêmicos (disponíveis em `artigos/`):

1. **Bayesian inference for Plackett-Luce ranking models** — base teórica do modelo de ranking
2. **Power EP Algorithm** — algoritmo de aproximação para inferência Bayesiana
3. **Bayesian Modelling and Analysis of Data** — fundamentos da modelagem Bayesiana
4. **Power EP** — extensão com α-divergence
5. **Modelo Mallows Concêntrico Top-k** — ranking parcial e concentração em torno do consenso
6. **Modelo Mallows Binomial** — variante binomial do modelo de Mallows
7. **Modelo Bayesiano Parcial** — inferência com rankings parcialmente observados
8. **Algoritmos Modelo Mallows (BT PL)** — Bradley-Terry e Plackett-Luce
9. **Neural RankGAM (LTR)** — Learning to Rank com redes neurais
